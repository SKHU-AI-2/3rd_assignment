# **모폴로지 연산 (Morphological Operations)**

1. **침식(Erosion)**: 노이즈 제거, 물체 경계 축소, 작은 구멍 제거.
2. **팽창(Dilation):** 구멍 채우기, 물체 연결, 경계 확장 (노이즈 증가 가능성).
3. **닫힘(Closing):** 팽창 후 침식. 구멍 메우기, 분리된 객체 연결.
4. **열림(Opening):** 침식 후 팽창. 노이즈 제거 및 객체 분리.


# **Color Space**

- **RGB**: 빛 혼합 비율로 색상 표현, 표현 불가능 영역 존재.
- **HSV**: 색상(Hue), 채도(Saturation), 명도(Value)로 구성. 독립적으로 제어 가능.

# **Edge Detection:**

이미지 밝기 변화 부분을 찾아 물체 경계를 추출


---
- Prewitt Operator: 간단하고 빠르나 명확한 밝기 변화가 필요.
- Sobel Operator: 중앙부 가중치 증가로 노이즈 제거 효과 우수.
- Chen-Frei Operator: 더 세밀한 경계 검출 가능하지만 계산량 증가.
